# RANK

```python
import numpy as np
np.random.seed(1)
data = pd.DataFrame({
    'name':['a','b','c','d','e'],
    'balance': np.random.randint(1,1000,5),
    'defaults': np.random.randint(1,15,5)
})
data['rank_balance']= data['balance'].rank(ascending=False) #ranking based on minimum value
data
```

```python
data['rank_defaults']= data['defaults'].rank() #rank based on minimum value
data
```

```python
data = pd.DataFrame({
    'name':['a','b','c','d','e'],
    'marks': [200,300,300,400,500]
})
data['rank_average'] = data['marks'].rank(ascending=False, method = 'average')
data['rank_min'] = data['marks'].rank(ascending=False,method='min')
data['rank_max'] = data['marks'].rank(ascending=False,method='max')
data['rank_dense'] = data['marks'].rank(ascending=False,method='dense')
data
```
