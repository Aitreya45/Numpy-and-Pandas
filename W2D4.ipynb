# EXERCISE

```python
#Q. Filter records for ['Sachin R Tendulkar','Virendra Sehwag','Rahul Dravid']
#Q. create a heatmap, where rows will have player names,columns will have Versus,values and equal average runs
```

```python
Players = ['Sachin R Tendulkar','Virender Sehwag','Rahul Dravid']
odi_subset = odi[odi['Player'].isin(Players)]
summary_long = odi_subset.groupby(['Player','Versus']).agg({'Runs':'mean'})
summary_wide = summary_long.unstack(level=1)
sns.heatmap(summary_wide,cmap='Blues')
```

```python
##Pivot Table
summary_pivot = pd.pivot_table(odi_subset,
                              index = 'Player',
                              columns = 'Versus',
                              values = 'Runs',
                              aggfunc = 'mean')
summary_pivot
```

```python
#stack: wide to long format
summary_pivot.stack().reset_index().rename(columns={0:'Avg Runs'}).head(5)
```
