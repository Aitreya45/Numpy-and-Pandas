# String Operation

```python
#calculate year wise total for Sachin R Tendulkar
odi['year'] = odi['MatchDate'].apply(lambda d: d[-4:])
sachin_rows = odi[odi['Player']== 'Sachin R Tendulkar']
sachin_years = sachin_rows.groupby(['year']).agg({'Runs':'sum'})
sachin_years
```

```python
odi['year'] = odi['MatchDate'].apply(lambda d: d[-4:])
rahul_rows = odi[odi['Player']== 'Rahul Dravid']
rahul_years = rahul_rows.groupby(['year']).agg({'Runs':'sum'})
rahul_years
```

# Joining

```python
#concat
df_new = pd.concat([sachin_years,rahul_years],axis=1,sort=False,join='outer')
df_new
```

```python
# Renaming columns
df_new = pd.concat([sachin_years,rahul_years],axis=1,sort=False,join='outer')
df_new.columns = ['Runs_sachin','Runs_Rahul']
df_new
```

```python
df_new = sachin_years.join(rahul_years,
                          lsuffix='_sachin',
                          rsuffix='_rahul',
                          how='outer')
df_new
```

# Merge


```python
sachin_new = sachin_years.reset_index()#moves index values as new cols
sachin_new
```

```python
rahul_new = rahul_years.reset_index()
rahul_new
```

```python
sachin_years.set_index('Runs')#move column to index
```

```python
df_merge = pd.merge(left=sachin_new,
                   right=rahul_new,
                   left_on = 'year',
                   right_on = 'year',
                   suffixes=('_sachin','_rahul'),
                   how='outer')
df_merge
```
